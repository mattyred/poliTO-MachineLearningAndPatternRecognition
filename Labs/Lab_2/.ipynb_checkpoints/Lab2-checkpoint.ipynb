# Lab 2 - iris dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
def mcol(row):
    return row.reshape(4, 1) # convert a row array into a column one
```

```python
def load(filename):
    label_numerical = {'Iris-setosa':0,
                       'Iris-versicolor':1,
                       'Iris-virginica':2}
    DList = [] # list of vertical arrays
    labelsList = [] # list of numerical labels
    with open(filename) as f:
        for line in f:
            try:
                attributes = np.array([float(i) for i in line.split(',')[0:4]])
                label = line.split(',')[-1].strip()
                DList.append(mcol(attributes)) 
                labelsList.append(label_numerical[label])
            except:
                pass
    return np.hstack(DList), np.array(labelsList, dtype = np.int32)
```

```python
features,labels = load('iris.csv')
# other way using sklearn...
```

```python
# Extract only features for Iris-setosa
print('Number of Iris-setosa: %d' % ((labels==0).sum()))
features_setosa = features[:,labels == 0]
```

```python
# Extract features for Iris-versicolor and Iris-virginica
features_versicolor = features[:,labels == 1]
features_virginica = features[:,labels == 2]
```

Now we can make an histogram with a feature on the x-axis and the normalized number of samples on the y-axis

```python
# iterate over the features
nameFeatures = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']
plt.figure()
for f in range(4):
    plt.hist(features_setosa[f], bins=20, density=True, alpha=0.4)
    plt.hist(features_versicolor[f], bins=20, density=True, alpha=0.4)
    plt.hist(features_virginica[f], bins=20, density=True, alpha=0.4)
    plt.legend(['setosa','versicolor','virginica'])
    plt.xlabel(nameFeatures[f])
    plt.show()
```

We can also plot a scatter plot with all the possible pairs of features

```python
nameFeatures = {0:'Sepal length', 
                1:'Sepal width',
                2:'Petal length',
                3:'Petal width'}
for i in range(4):
    for j in range(4):
        plt.figure()
        if i != j:
            plt.scatter(features_setosa[i], features_setosa[j], label = 'setosa')
            plt.scatter(features_versicolor[i], features_versicolor[j], label = 'versicolor')
            plt.scatter(features_virginica[i], features_virginica[j], label = 'virginica')
            plt.xlabel(nameFeatures[i])
            plt.ylabel(nameFeatures[j])
            plt.legend()
    plt.show()
```
